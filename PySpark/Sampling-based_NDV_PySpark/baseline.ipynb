```python
import os
memory = '200g'
pyspark_submit_args = ' --driver-memory ' + memory  +' pyspark-shell'+" --num-executors 15 --executor-cores 8"
os.environ["PYSPARK_SUBMIT_ARGS"] = pyspark_submit_args
import numpy as np
from collections import Counter
#import hyperloglog as hll
#from HyperLogLog import Hyperloglog as hll
#from HyperLogLog import Countsketch
import time
import pandas as pd
import pickle
import math
import random
import findspark
findspark.init()
from pyspark import SparkContext
sc=SparkContext(master="spark://Master:17077",appName="MapReduce NDV")
#sc=SparkContext(appName="test")
sc._conf.set("spark.driver.maxResultSize", '50g')
```

```python
def GEE(NDV,f1,q):
    return f1*(math.sqrt(1/q)-1)+NDV
def Chao(NDV,f1,f2):
    return NDV+f1*(f1-1)/2/(f2+1)
def Shlosser(NDV,f1,f,q):
    Ef=0
    Exf=0
    for (key,value) in f:
        Ef+=(1-q)**key*value
        Exf+=q*key*(1-q)**(key-1)*value
    return NDV+Ef/Exf*f1
def CL1(NDV,f1,f):
    tm=0
    n=0
    for (key,value) in f:
        n+=key*value
    for (key,value) in f:
        tm+=key*(key-1)*value
    tm=max(0, NDV*tm/(1-f1/n)/(n*n-n-1))
    return (NDV+f1*tm)/(1-f1/n)
```

```python
file_path='/user/poi/poi-50-1kww/poi-50-1kww/'
file1rdd=sc.textFile(file_path)
begin=time.time()
n=file1rdd.count()
end=time.time()
print("count n time:",end-begin)
print("n:",n)
q=0.01
begin=time.time()
part_dict = file1rdd.map(lambda x:(x,1)).reduceByKey(lambda a,b:a+b)\
    .map(lambda d:(d[1],1)).reduceByKey(lambda e,f:e+f).collect()
NDV=0
f1=0
f2=0
for (key,value) in part_dict:
    NDV+=value
    if key==1:
        f1=value
    if key==2:
        f2=value
DGEE=GEE(NDV,f1,q)
DChao=Chao(NDV,f1,f2)
Dshlosser=Shlosser(NDV,f1,part_dict,q)
DCL1=CL1(NDV,f1,part_dict)
end=time.time()
print("MapReduceSampleTime:",end-begin)
print("GEE:",DGEE)
print("Chao:",DChao)
print("Shlosser:",Dshlosser)
print("DCL1:",DCL1)
print("d:",NDV)
print("f1:",f1)
print("f2:",f2)
```
